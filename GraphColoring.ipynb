```python
def greedy_color(adj):
    # Welshâ€“Powell ordering: sort vertices by degree
    order = sorted(adj, key=lambda v: len(adj[v]), reverse=True)
    color = {}

    for v in order:
        used = {color[n] for n in adj[v] if n in color}
        # pick the smallest color 1..4 not used by it's neighbors
        for c in range(1, 5):
            if c not in used:
                color[v] = c
                break
        else:
            raise ValueError("Needs more than 4 colors for this ordering.")
    return color
```

```python
adjacency = {
    # Florida touches: AL, GA
    "FL": {"AL", "GA"},
    # Georgia touches: FL, AL, TN, NC, SC
    "GA": {"FL", "AL", "TN", "NC", "SC"},
    # Alabama touches: FL, GA, TN, MS
    "AL": {"FL", "GA", "TN", "MS"},
    # Tennessee touches: GA, AL, MS, AR, MO, KY, NC
    "TN": {"GA", "AL", "MS", "AR", "MO", "KY", "NC"},
    # North Carolina touches: GA, TN, SC, VA
    "NC": {"GA", "TN", "SC", "VA"},
    # South Carolina touches: GA, NC
    "SC": {"GA", "NC"},
    # Mississippi touches: AL, TN, AR, LA
    "MS": {"AL", "TN", "AR", "LA"},
    # Arkansas touches: TN, MS, MO, LA, TX, OK
    "AR": {"TN", "MS", "MO", "LA", "TX", "OK"},
    # Louisiana touches: MS, AR, TX
    "LA": {"MS", "AR", "TX"},
    # Texas touches: NM, OK, AR, LA  (NM omitted from this subset)
    "TX": {"OK", "AR", "LA"},
    # Oklahoma touches: CO, KS, MO, AR, TX, NM (limited subset here)
    "OK": {"MO", "AR", "TX"},
    # Missouri touches: OK, AR, TN, KY
    "MO": {"OK", "AR", "TN", "KY"},
    # Kentucky touches: MO, TN, VA
    "KY": {"MO", "TN", "VA"},
    # Virginia touches: NC, KY
    "VA": {"NC", "KY"},
}
```

```python
# Runs the greedy coloring
coloring = greedy_color(adjacency)

# Prints lookup tables
print("Lookup Table: State = Color")
for s in sorted(coloring):
    print(f"{s:>12} = {coloring[s]}")
```
