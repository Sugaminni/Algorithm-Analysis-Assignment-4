```python
import heapq

# Performs dijkstra's algorithm
def dijkstra(adj, source='A'):
    dist = {v: float('inf') for v in adj} # Makes all distances infinity
    dist[source] = 0.0 # Distance to source node is always at 0
    pq = [(0.0, source)]
    while pq: # Processes all nodes
        d, u = heapq.heappop(pq)
        if d > dist[u]:
            continue
        for v, w in adj[u]:
            nd = d + w
            if nd < dist[v]:
                dist[v] = nd
                heapq.heappush(pq, (nd, v))
    return dist

# Graph from the figure shown in Q4
edges = [
    ('A','C',3),  ('C','B',15), ('A','B',12),
    ('A','E',6),  ('B','E',5),  ('E','G',22),
    ('B','D',5),  ('D','F',4),  ('F','H',10),
    ('D','H',19), ('H','G',4),  ('C','F',18),
]

# Builds adjacency list from edge list
adj = {}
for u, v, w in edges:
    adj.setdefault(u, []).append((v, w))
    adj.setdefault(v, []).append((u, w))

# Prints the shortest distances from A to all other nodes
dist = dijkstra(adj, 'A')
print("Shortest distances from A:")
for node in sorted(dist):
    print(f"{node}: {dist[node]}")
```
